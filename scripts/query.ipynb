```python
import pyarrow.parquet as pq
import pandas as pd

df = pq.read_table('../data/chat.parquet').to_pandas()
```

```python
pd.set_option('display.max_rows', 300)
```

```python
query = df.query("timestamp >= '2022-09-01' and timestamp <= '2023-07-01' and sender == 'Grana' and newline_count > 2").drop([3198, 3759, 4232, 4266, 4653, 4874, 6329, 6353, 7297])
display(query)
```

```python
from datetime import datetime, timedelta
from unidecode import unidecode
import re

season_start = datetime.strptime('2023-09-05', '%Y-%m-%d')
season_end = datetime.strptime('2024-06-27', '%Y-%m-%d')
current_date = season_start

original_dates = []
real_dates = []
descriptions = []
summaries = []

session_iterator = query.iterrows()

try:
    while current_date <= season_end or session:
        session = next(session_iterator)[1]
        plain_description = unidecode(session['message']).lower()

        # If it's a summary we do not append a new session
        if re.match(r'resumen', plain_description):
            summaries[-1] = session['message'].strip()
            continue

        original_dates.append(session['timestamp'].strftime('%Y-%m-%d'))
        real_dates.append(current_date.strftime('%Y-%m-%d'))
        descriptions.append(session['message'].strip())
        summaries.append("")

        current_date = current_date + timedelta(days=2) if current_date.weekday() == 1 else current_date + timedelta(days=5)
except StopIteration as e:
    print('Ran out of sessions')

new_schedule = pd.DataFrame({'fecha': real_dates, 'fecha original': original_dates, 'descripcion': descriptions, 'resumen': summaries})
display(new_schedule)
```

```python
new_schedule.to_csv('../out/2023-2024_schedule.csv')
```
